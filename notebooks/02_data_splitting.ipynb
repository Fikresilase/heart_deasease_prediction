# 02 Data Splitting

Train/test split for the cleaned heart disease dataset.


```python
import pandas as pd
from pathlib import Path
```

```python
processed_dir = Path("../data/processed")
data_path = processed_dir / "heart_cleaned.csv"

df = pd.read_csv(data_path)
df.head()
```

```python
train_parts = []
test_parts = []

for _, group in df.groupby("target", sort=True):
    test_part = group.sample(frac=0.20, random_state=42)
    train_part = group.drop(test_part.index)
    train_parts.append(train_part)
    test_parts.append(test_part)

heart_train = pd.concat(train_parts).sample(frac=1, random_state=42).reset_index(drop=True)
heart_test = pd.concat(test_parts).sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
X_train = heart_train.drop(columns=["target"])
y_train = heart_train["target"]

X_test = heart_test.drop(columns=["target"])
y_test = heart_test["target"]
```

```python
X_train.to_csv(processed_dir / "X_train.csv", index=False)
X_test.to_csv(processed_dir / "X_test.csv", index=False)
y_train.to_csv(processed_dir / "y_train.csv", index=False)
y_test.to_csv(processed_dir / "y_test.csv", index=False)
heart_train.to_csv(processed_dir / "heart_train.csv", index=False)
heart_test.to_csv(processed_dir / "heart_test.csv", index=False)
```

```python
split_summary = pd.DataFrame({
    "dataset": ["full", "train", "test"],
    "rows": [len(df), len(heart_train), len(heart_test)],
    "target_0": [int((df["target"] == 0).sum()), int((y_train == 0).sum()), int((y_test == 0).sum())],
    "target_1": [int((df["target"] == 1).sum()), int((y_train == 1).sum()), int((y_test == 1).sum())]
})

split_summary
```
